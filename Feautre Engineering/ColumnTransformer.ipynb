```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("../datasets/CarSpeedData..csv")
```

```python
df.rename(columns={"Unnamed: 1":"Speed"},inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop("has_driving_license",axis='columns'),df["has_driving_license"],test_size=0.2)
x_train.shape,x_test.shape
```

```python
from sklearn.compose import ColumnTransformer
transformer= ColumnTransformer(transformers=[
    ('t1',SimpleImputer(),["Speed"]),
    ('t2',OneHotEncoder(sparse_output=False,drop="first"),["Gender","City"]),
    ('t3',OrdinalEncoder(categories=[['low','high']]),['Average_speed'])
],remainder='passthrough')
```

```python
transformer.fit_transform(x_train).shape
transformer.transform(x_test).shape
```
