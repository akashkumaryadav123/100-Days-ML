```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../datasets/Wine dataset.csv",usecols=[0,1,2])
```

```python
df
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('class',axis=1),df["class"],test_size=0.20,random_state=0)
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)

x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)


```

```python
x_test_scaled.sample(10)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
sns.kdeplot(x_train["Malic acid"], fill=True, color="grey", label="Estimated PDF",ax=axes[0])
sns.kdeplot(x_train_scaled["Malic acid"], fill=True, color="red", label="Estimated PDF",ax=axes[1])
axes[0].set_title("Before scaling")
axes[1].set_title("After scaling")
```
