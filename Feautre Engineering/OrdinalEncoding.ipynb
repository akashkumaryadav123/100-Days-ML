```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

```

```python
df=pd.read_csv("../datasets/customer.csv")
```

```python
df.sample(10
         )
```

```python
df=df.iloc[:,2:]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:2],df["purchased"],test_size=0.2,random_state=0)
```

```python
x_train.shape , x_test.shape
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[["Poor","Average","Good"],["School","UG","PG"]])
oe.fit(x_train)
x_train_oe=oe.transform(x_train)
x_test_oe=oe.transform(x_test)
```

```python
x_train_oe=pd.DataFrame(x_train_oe,columns=x_train.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(y_train)
y_train_le=le.transform(y_train)
y_test_le=le.transform(y_test)
```
