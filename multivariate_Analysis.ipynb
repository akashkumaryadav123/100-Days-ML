```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv")
```

```python
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
```

```python
df.head(20)
```

# ScatterPlot(Numerical-Numerical)

```python
sns.scatterplot(x=df["Age"],y=df["Survived"])
```

# DistPlot (N-C)

```python
sns.kdeplot(df[df["Survived"]==0]["Age"],label="Did not survived")
sns.kdeplot(df[df["Survived"]==1]["Age"],label="Survived")
plt.legend()
```

```python
df.corr(numeric_only=True)
```

# Heatmap(C-C)

```python
df.head(2)
```

```python
sns.heatmap(pd.crosstab(df["Sex"],df["Survived"]))
```

```python
df.groupby("Sex")["Survived"].mean() * 100
```

```python
df["Sex"].value_counts().plot(kind="pie",autopct="%2.f")
```

# Clusterplot (C-C)

```python
# here we usaully see relation between different group in heirracy fformat
sns.clustermap(pd.crosstab(df["Parch"],df["Survived"]))
```
